```python
plt.subplot(1,2,1)
# nrows 
# n cols 
# plot number

plt.plot(data['fixed acidity'])
plt.title('Fixed Acidity')
plt.xlabel('number of wines')


plt.subplot(1,2,2)
plt.plot(data['pH'])
plt.title('pH')

# for common title
plt.suptitle('Fixed Acidity and Ph')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(data['fixed acidity'], color='blue')
plt.subplot(1,2,2)
plt.plot(data['pH'], color='red')

# for common title
plt.suptitle('Fixed Acidity and Ph')
plt.show()
```

```python
plt.figure(figsize = (10,5)) # units in inches 

plt.subplot(1,2,1)
plt.plot(data['fixed acidity'], color='blue')
plt.subplot(1,2,2)
plt.plot(data['pH'], color='red')

# for common title
plt.suptitle('Fixed Acidity and Ph')
plt.show()
```

```python
plt.scatter(x=data['fixed acidity'], y=data['pH'])
plt.ylabel('pH ')
plt.xlabel('fixed acidity')
plt.show()
```

```python
plt.scatter(x=data['fixed acidity'], y=data['pH'], alpha=0.2)
plt.ylabel('pH ')
plt.xlabel('fixed acidity')
plt.xticks([6,7,8,9,10,11,12])
plt.show()
```

```python
plt.scatter(x=data['fixed acidity'], y=data['pH'], alpha=0.2, color='red')
plt.ylabel('pH ')
plt.xlabel('fixed acidity')
plt.xticks()
plt.show()
```
